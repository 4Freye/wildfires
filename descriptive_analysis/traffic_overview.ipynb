Top Outflow & Inflow counties during Kincaid fire:

```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt
import json

with open('../data/cali_counties_lat_long_dict.json') as json_file:
    cali_counties_lat_long_dict = json.load(json_file)
```

```python
outflow_ts = pd.read_csv('../data/california_jul_nov_2019.csv').groupby(['geoid_o','date']).agg({'visitor_flows':'sum'}).reset_index()
outflow_ts['date'] = pd.to_datetime(outflow_ts.date)
outflow_ts['origin_county'] = outflow_ts.geoid_o.astype(str).map(cali_counties_lat_long_dict)
outflow_ts = outflow_ts.drop('geoid_o', axis=1).pivot(values='visitor_flows',index='date',columns='origin_county')

mc_decomposed = seasonal_decompose(outflow_ts[['Monterey County']], model='multiplicative')
sonoma_decomposed = seasonal_decompose(outflow_ts[['Sonoma County']], model='multiplicative')
import matplotlib.pyplot as plt
monterey = pd.Series(mc_decomposed._trend * mc_decomposed._resid, name = 'Monterey County')
sonoma = pd.Series(sonoma_decomposed._trend * sonoma_decomposed._resid, name = 'Sonoma County')
ts_df = pd.concat([monterey, sonoma], axis=1)
ts_df.plot(title='Daily Visitor Outflows from Sonoma and Monterey County')
# Convert the date string to a datetime object
date_str = '2019-10-23'
date = pd.to_datetime(date_str)

# Get the x-axis position for the date
x_pos = date.toordinal()

# Add a vertical line at the specified date
plt.axvline(x=date, color='red', dashes= (2,1))
# Add text to the vertical line
plt.text(date, plt.ylim()[1]/1.05, 'start of Kincaid Fire ', color='red', ha='right')

# Display the plot
plt.show()
```
